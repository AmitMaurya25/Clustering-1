```python
Clustering algorithms are unsupervised learning techniques that group similar data points together based on certain criteria. 
There are several types of clustering algorithms, and they can be broadly categorized into the following:

    Hierarchical Clustering:
        Approach: Hierarchical clustering builds a hierarchy of clusters. It starts with each data point as its own cluster and iteratively 
        merges or splits clusters until a certain criterion is met.
        Assumptions: Assumes a hierarchy of relationships between data points.
        Example: Agglomerative Hierarchical Clustering, Divisive Hierarchical Clustering.

    
```

```python
# Example using Agglomerative Hierarchical Clustering
from sklearn.datasets import make_blobs
from sklearn.cluster import AgglomerativeClustering
import matplotlib.pyplot as plt

X, _ = make_blobs(n_samples=300, centers=3, random_state=42)

hierarchical = AgglomerativeClustering(n_clusters=3)
labels = hierarchical.fit_predict(X)

plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', edgecolor='k', s=50)
plt.title('Agglomerative Hierarchical Clustering')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```

```python
Partitioning Clustering:

    Approach: Partitioning clustering involves dividing the dataset into non-overlapping subsets (clusters). Each data point belongs to only one cluster.
    Assumptions: Assumes that the number of clusters is known in advance and the partitions are formed based on certain criteria.
    Example: K-Means Clustering.
```

```python
# Example using K-Means Clustering
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

X, _ = make_blobs(n_samples=300, centers=3, random_state=42)

kmeans = KMeans(n_clusters=3, random_state=42)
labels = kmeans.fit_predict(X)

plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', edgecolor='k', s=50)
plt.title('K-Means Clustering')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```

```python
Density-Based Clustering:

    Approach: Density-based clustering groups together data points that are close to each other in high-density regions and separates points
    in low-density regions.
    Assumptions: Assumes that clusters have varying densities and can have arbitrary shapes.
    Example: DBSCAN (Density-Based Spatial Clustering of Applications with Noise).
```

```python
# Example using DBSCAN
from sklearn.datasets import make_moons
from sklearn.cluster import DBSCAN
import matplotlib.pyplot as plt

X, _ = make_moons(n_samples=200, noise=0.05, random_state=42)

dbscan = DBSCAN(eps=0.3, min_samples=5)
labels = dbscan.fit_predict(X)

plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', edgecolor='k', s=50)
plt.title('DBSCAN Clustering')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```

```python
Model-Based Clustering:

    Approach: Model-based clustering assumes that the data is generated from a mixture of underlying probability distributions.
    It seeks to identify the parameters of these distributions.
    Assumptions: Assumes that the data can be modeled by a specific statistical distribution (e.g., Gaussian distribution).
    Example: Gaussian Mixture Models (GMM).
```

```python
# Example using Gaussian Mixture Models (GMM)
from sklearn.datasets import make_blobs
from sklearn.mixture import GaussianMixture
import matplotlib.pyplot as plt

X, _ = make_blobs(n_samples=300, centers=3, random_state=42)

gmm = GaussianMixture(n_components=3, random_state=42)
labels = gmm.fit_predict(X)

plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', edgecolor='k', s=50)
plt.title('Gaussian Mixture Model (GMM) Clustering')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```

## Qestion 2 --------------------------------------------------------------------------------------------------------------

```python
K-means clustering is a partitioning algorithm that divides a dataset into K distinct, non-overlapping subsets (clusters).
It is one of the most widely used clustering algorithms. The algorithm works iteratively to assign each data point to one of 
K clusters based on their features, aiming to minimize the within-cluster sum of squared distances.
How K-means Works:

    Initialization:
        Randomly select K initial cluster centroids, where K is the number of desired clusters.

    Assignment:
        Assign each data point to the cluster whose centroid is closest, typically using Euclidean distance.

    Update Centroids:
        Recalculate the centroids of each cluster based on the mean of the data points assigned to that cluster.

    Repeat:
        Repeat steps 2 and 3 until convergence. Convergence occurs when the assignments of data points to clusters no longer change or change minimally.

Code Example:

Let's demonstrate K-means clustering using scikit-learn on a synthetic dataset:
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

# Generate synthetic data with three blobs
X, _ = make_blobs(n_samples=300, centers=3, random_state=42)

# Apply K-means clustering with K=3
kmeans = KMeans(n_clusters=3, random_state=42)
labels = kmeans.fit_predict(X)

# Visualize the clustering results
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', edgecolor='k', s=50)
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], c='red', marker='X', s=200, label='Centroids')
plt.title('K-Means Clustering')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.show()
```

## Qestion 3 --------------------------------------------------------------------------------------------------------------

```python
Advantages of K-means Clustering:

    Simplicity and Speed:
        K-means is computationally efficient and relatively simple to implement.
        Well-suited for large datasets and high-dimensional data.

    Scalability:
        Performs well on large datasets and is scalable to a large number of samples.

    Interpretability:
        The clusters formed by K-means are easy to interpret and visualize.

    Global Optimality:
        Despite its sensitivity to initialization, K-means converges to a local minimum, which is often a global minimum.

    Applicability:
        Widely used in practice across various domains due to its simplicity and effectiveness.

Limitations of K-means Clustering:

    Assumption of Spherical Clusters:
        K-means assumes that clusters are spherical and equally sized, making it less effective for clusters with non-uniform shapes and sizes.

    Sensitivity to Initial Centroids:
        The choice of initial cluster centroids can impact the final clustering result. K-means may converge to different solutions based on the initialization.

    Fixed Number of Clusters (K):
        Requires specifying the number of clusters (K) in advance, which may not be known in real-world applications.

    Sensitive to Outliers:
        K-means is sensitive to outliers, and their presence can significantly impact the centroids and cluster assignments.

    Euclidean Distance Metric:
        Relies on the Euclidean distance metric, which may not be suitable for all types of data (e.g., categorical data or data with varying scales).

    Non-Convex Clusters:
        Ineffective for datasets with clusters of non-convex shapes, as K-means tends to produce circular/spherical clusters.

Code Example:

Let's illustrate the sensitivity to initial centroids using the same synthetic dataset as before:
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

# Generate synthetic data with three blobs
X, _ = make_blobs(n_samples=300, centers=3, random_state=42)

# Apply K-means clustering with different initializations
kmeans1 = KMeans(n_clusters=3, random_state=42)
labels1 = kmeans1.fit_predict(X)

kmeans2 = KMeans(n_clusters=3, random_state=0)  # Different random_state
labels2 = kmeans2.fit_predict(X)

# Visualize the clustering results with different initializations
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.scatter(X[:, 0], X[:, 1], c=labels1, cmap='viridis', edgecolor='k', s=50)
plt.scatter(kmeans1.cluster_centers_[:, 0], kmeans1.cluster_centers_[:, 1], c='red', marker='X', s=200, label='Centroids')
plt.title('K-Means Clustering (Initialization 1)')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()

plt.subplot(1, 2, 2)
plt.scatter(X[:, 0], X[:, 1], c=labels2, cmap='viridis', edgecolor='k', s=50)
plt.scatter(kmeans2.cluster_centers_[:, 0], kmeans2.cluster_centers_[:, 1], c='red', marker='X', s=200, label='Centroids')
plt.title('K-Means Clustering (Initialization 2)')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()

plt.tight_layout()
plt.show()
```

## Qestion 4 --------------------------------------------------------------------------------------------------------------

```python
Determining the optimal number of clusters (K) in K-means clustering is a crucial step, and several methods can help in making an informed choice.
Here are some common methods:
1. Elbow Method:

The Elbow Method involves running K-means clustering on the dataset for a range of values of K and plotting the within-cluster sum of squared 
distances (inertia) against K. The "elbow" in the plot is often considered the optimal K, where adding more clusters doesn't significantly reduce 
inertia.
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

# Generate synthetic data with varying number of blobs
X, _ = make_blobs(n_samples=300, centers=4, random_state=42)

# Use the Elbow Method to determine optimal K
inertias = []
possible_k_values = range(1, 11)

for k in possible_k_values:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X)
    inertias.append(kmeans.inertia_)

# Plot the Elbow Method curve
plt.plot(possible_k_values, inertias, marker='o')
plt.title('Elbow Method for Optimal K')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Inertia (Within-Cluster Sum of Squared Distances)')
plt.show()
```

```python
2. Silhouette Score:

The Silhouette Score measures how well-separated the clusters are. It calculates the mean silhouette coefficient for each instance and ranges 
from -1 to 1, where a higher score indicates better-defined clusters.
```

```python
from sklearn.metrics import silhouette_score

silhouette_scores = []

for k in possible_k_values:
    kmeans = KMeans(n_clusters=k, random_state=42)
    labels = kmeans.fit_predict(X)
    silhouette_scores.append(silhouette_score(X, labels))

# Plot the Silhouette Score curve
plt.plot(possible_k_values, silhouette_scores, marker='o')
plt.title('Silhouette Score for Optimal K')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Silhouette Score')
plt.show()
```

```python
3. Gap Statistics:

Gap Statistics compare the within-cluster sum of squared distances of the clustering solution with the sum of squared distances in a 
reference null distribution. A larger gap indicates a better clustering solution.
```

```python
from sklearn.cluster import KMeans
from gap_statistic import OptimalK

optimal_k = OptimalK(parallel_backend='joblib')
n_clusters = optimal_k(X, cluster_array=np.arange(1, 11))
print(f'Optimal number of clusters (Gap Statistics): {n_clusters}')
```

```python
K-means clustering has various applications across different domains due to its simplicity and effectiveness. Here are some real-world scenarios 
where K-means clustering has been successfully applied:
1. Image Compression:

K-means clustering is used for image compression by reducing the number of colors in an image. Each pixel is assigned to the cluster centroid, 
resulting in a compressed image with a reduced color palette.
```

```python
from sklearn.cluster import KMeans
from sklearn.utils import shuffle
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

# Load an image
image_path = 'path_to_image.jpg'
original_image = Image.open(image_path)

# Convert the image to a numpy array
image_array = np.array(original_image)

# Reshape the array to 2D
w, h, d = image_array.shape
image_2d = image_array.reshape((w * h, d))

# Perform K-means clustering
kmeans = KMeans(n_clusters=16, random_state=42)
kmeans.fit(image_2d)
labels = kmeans.predict(image_2d)
cluster_centers = kmeans.cluster_centers_.astype(int)
compressed_image = cluster_centers[labels].reshape((w, h, d))

# Display the original and compressed images
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(image_array)
plt.title('Original Image')

plt.subplot(1, 2, 2)
plt.imshow(compressed_image)
plt.title('Compressed Image')

plt.show()
```
